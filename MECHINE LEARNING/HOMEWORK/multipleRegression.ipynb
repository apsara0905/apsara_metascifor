```python
import matplotlib.pyplot as plt
import pandas as pd
import statsmodels.api as sm
import numpy as np

# Create a DataFrame with multiple predictors
data = {
    'Actual Values': [10.84, 9.01, 11.41, 4.20, 5.75, 29.75, 6.90, 7.15, 7.23],
    'Predictor 1': [1.2, 2.3, 3.1, 4.2, 5.5, 6.6, 7.0, 8.1, 9.3],
    'Predictor 2': [3.3, 4.4, 5.1, 6.5, 7.2, 8.3, 9.1, 10.2, 11.0]
}

df = pd.DataFrame(data)

# Add a constant to the predictors (intercept term)
X = sm.add_constant(df[['Predictor 1', 'Predictor 2']])
y = df['Actual Values']

# Fit the multiple regression model
model = sm.OLS(y, X).fit()
df['Predicted Values'] = model.predict(X)

# Plot Actual vs. Predicted Values
plt.figure(figsize=(10, 6))
plt.scatter(df['Actual Values'], df['Predicted Values'], color='blue', marker='o')
plt.plot([df['Actual Values'].min(), df['Actual Values'].max()],
         [df['Actual Values'].min(), df['Actual Values'].max()],
         linestyle='-', color='blue')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Actual vs. Predicted Values')
plt.grid(True)

# Display the plot
plt.show()
```
